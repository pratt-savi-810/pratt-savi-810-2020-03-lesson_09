#### Let's grab a road network around Pratt University

````python
# source: https://github.com/gboeing/osmnx-examples/tree/master/notebooks

pratt_network = ox.graph_from_address(
    address='200 Willoughby Ave, Brooklyn, NY 11205',
    distance=1000,  # meters
    distance_type='network',
    retain_all=True,
    truncate_by_edge=False,
    simplify=False,
)
````

#### Let's parse those out into a set of GeoDataFrames

````python
# source: https://github.com/gboeing/osmnx-examples/tree/master/notebooks
graph_gdfs = ox.graph_to_gdfs(pratt_network, edges=True, nodes=True) 
````

#### And pull out the intersections and the streets

````python
pratt_intersections = graph_gdfs[0]
pratt_streets = graph_gdfs[1]
````

#### Let's run this code to get our `address` DataFrame from last lesson. 

````python
from geopy.geocoders import Nominatim
from shapely.geometry import Point  # importing the Point class
from geopandas import GeoDataFrame

data = {
    'address': [
        '175 5th Avenue NYC',
        '200 Willoughby Ave, Brooklyn, NY 11205',
        '315 Bowery, NY, NY',
    ],
    'name': [
        'Flatiron Bldg',
        'Pratt',
        "CBGB's"
    ]
}

geolocator = Nominatim(user_agent='pratt_geospatial')

df = pd.DataFrame(data, columns=data.keys())

df['geocode'] = df['address'].apply(geolocator.geocode)  

df['geometry'] = df['geocode'].apply(
    lambda x: 
        Point(x.longitude, x.latitude)
) # create a geometry column

address = GeoDataFrame(df, geometry='geometry', crs='epsg:4326')
````

#### Let's grab just the pratt address from our `address` GeoDataFrame

````python
pratt_point = address[address['address'] == '200 Willoughby Ave, Brooklyn, NY 11205']
````

````python
pratt_point
````

#### Let's overlay our Radial Buffer and our Network Analysis 1km output. 

````python
# source: https://github.com/gboeing/osmnx-examples/tree/master/notebooks

fig, ax = plt.subplots(figsize=(12, 12))

pratt_point.to_crs(epsg=2263).buffer(3280.84).plot(ax=ax)
pratt_streets.to_crs(epsg=2263).plot(ax=ax, color='red')
pratt_intersections.to_crs(epsg=2263).plot(ax=ax, color='pink')

pratt_point.to_crs(epsg=2263).plot(ax=ax, color='black', marker='X', markersize=5000)

plt.title("Radial vs. Network Buffer")

plt.show();
````

# Let's use a smaller buffer and "Clip" our Streets Data

#### Copy our `pratt_point`

````python
circle = pratt_point.copy()
````

#### Plot our `circle`

````python
circle.plot()
````

#### Change the CRS to NY State Plane


````python
circle.to_crs(epsg=2263, inplace=True)
````

#### Set it's geometry column to a buffer of 1000 feet

````python
circle['geometry'] = circle.buffer(1000)
````

#### We'll actually set back to WGS 84, not good in practice

````python
circle.to_crs(epsg=4326, inplace=True)
````

#### Plot

````python
fig, ax = plt.subplots(figsize=(12, 12))

circle.plot(ax=ax)
pratt_streets.plot(ax=ax, color='red')
#intersections.plot(ax=ax, color='pink')

plt.title('Pratt Streets with Smaller Buffer')

plt.show();
````

#### Set the Map Title by applying to Axes with `ax.set_title( )`

````python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 12))

print(ax)

ax1, ax2, ax3 = ax  # unpacking our list

nta_gdf.plot(ax=ax1)
ax1.set_xlim(980_000, 1_005_000)   # these underscores are like the thousand-separator 
ax1.set_ylim(190_000, 210_000)
ax1.set_title('AXES Map Title')

nta_gdf.plot(ax=ax2)
ax2.set_xlim(980_000, 1_005_000)   # these underscores are like the thousand-separator 
ax2.set_ylim(190_000, 210_000)
ax2.set_title('AXES Map Title TWWWWOOO')

nta_gdf.plot(ax=ax3)
ax3.set_xlim(980_000, 1_005_000)   # these underscores are like the thousand-separator 
ax3.set_ylim(190_000, 210_000)
ax3.set_title('AXES Map Title 3  THREEEE');
````
